# Problem # A

## A-1

```python
#Answer for A-1
import pandas as pd
import numpy as np
import matplotlib as plt

df = pd.read_excel('HW3_Data_A.xlsx')   # read the  Excel file and create the data frame
display(df.sample(15))                  # Display the random selectly rows using sample method
```

## A-2

```python
#Answer for A-2
headers = [header for header in df.columns]
rowLabels = [label for label in df.index]
print(headers)
print(rowLabels)

print(f"Number of headers : {len(headers)}")
print(f"Number of columns : {len(rowLabels)}")

print("*"*100)
print(f"Number of null rows \n {df.shape[0] - df.count(axis=0)}")  # shape gives the numbers of total count and df.count(axis=1) give non null values
```

## A-3

```python
#Answer for A-3
print("Summary of the numerical coulmns")
display(df.describe())

print("Summary of the Non numeric coulms")
cat_columns = df.select_dtypes(include='object').columns
display(df[cat_columns].describe())
```

## A-4

```python
#Answer for A-4
selected_rows = df['make'] == 'toyota'
new_df = df.loc[selected_rows,:]
print(f"number of rows in the data that corresponds to \"toyota\" car : {len(new_df.index)}")

display(new_df.describe(include='all'))
```

## A-5

```python
#Answer for A-5
mask = (df['fuel-type'] == "gas") & (df['horsepower'] < 130) & (df['horsepower'] >100)
new_df = df.loc[mask,:]

priceRnage = new_df["price"].max() - new_df['price'].min()

print(f"The range of the price is {priceRnage}")
```

## A-6

```python
#Answer for A-6
selected_rows = (df['length'] <= 170)
ndf = df.loc[selected_rows,'num-of-doors']
#print(ndf.value_counts())

proportion = ndf.value_counts()[0]/len(ndf.index)*100

print(f"proportion of cars having \"two\" doors / length greater than or equal to 170 is : {proportion: 0.2f}%")
```

## A-7

```python
#Answer for A-7
df.sort_values(by=['price'],inplace=True,ascending=True)
display(df.head(15))

new_df = df.iloc[:15,:]
selected_cols = new_df.select_dtypes(include = 'object').columns

display(new_df[selected_cols].describe())
```

## A-8

```python
#Answer for A-8
def changeCol(value):
    if value == "std":
        return "standard"
    else:
        return value
    
df['aspiration'] = df['aspiration'].apply(changeCol)
df.loc[:,['length','width','height']] = df.loc[:,['length','width','height']].applymap(lambda x : int(round(x))) 

display(df.loc[:,['aspiration','length','width','height']])
```

## A-9

```python
#Answer for A-9
df['city-kpl'] = df['city-mpg'].apply(lambda x: 0.425 * x)
df['highway-kpl'] = df['highway-mpg'].apply(lambda x: 0.425 * x)
df
```

# Problem # B

## B-1

```python
#Answer for B-1
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("HW3_Data_B.csv")

columns = df.columns

for c in columns:
    plt.figure()
    sns.histplot(x=c,bins=10,data=df)
    plt.show()
```

## B-2

```python
#Answer for B-2
plt.figure()
sns.histplot(x='loan_amount',bins=10,data=df)
plt.show()
#   Loan_amount
'''
    Mode = (0.2 - 0.3) x 10^ 6
    This not follow a normal distrubution. It is skewed to the left
'''
sns.histplot(x='rate_of_interest',bins=10,data=df)
plt.show()

#   Rate of interest
'''
    Mode = 3.9 - 4.7
    This is close to a normal distribution.It is mostly symmetric about the mean
'''

sns.histplot(x='loan_type',bins=10,data=df)
plt.show()
#   Loan_type
'''
    Mode = type1
    This also skewd to the left. 
'''

sns.histplot(x='property_value',bins=10,data=df)
plt.show()
#   Property_value
'''
    Mode  = (0.2 -0.3) x 10 ^ 6
    This close to a noraml distribution. But there is a some skewness to the left side
'''
```
